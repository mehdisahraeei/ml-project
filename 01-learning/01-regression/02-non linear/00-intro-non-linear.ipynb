# Non Linear Regression
- y = x3 + x2 + x + 1
- y = Log(x^2)

<h2 id="importing_libraries">Importing required libraries</h2>


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### first plot

```python
x = np.arange(-5.0, 5.0, 0.1)

# y = np.power(x,3)
y = 1*(x**3) + 1*(x**2) + 1*x + 3 # y = x^3 + x^2 + x^1 + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata,  'go')
plt.plot(x,y, 'r') 

plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Some non-linear examples


### Quadratic
- Y = x^2 and x^3 and ...

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata,  'go')
plt.plot(x,y, 'r') 

plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Exponential


- Y = a+bc^x
- exp() method returns E raised to the power of x (E^x) | E(approximately) = 2.718282

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = np.exp(X)

plt.plot(X,Y,'b-') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Logarithmic
- y = log(x)

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = np.log(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
# invalid value encountered in log: beacuse of having negative values inside the log
```

### Sigmoidal/Logistic
- is like S curve
- for example:
$$ Y = a + \frac{b}{1+ c^{(X-d)}}$$

```python
X = np.arange(-15.0, 15.0, 0.1)
Y = 1-4/(1+np.power(3, X-2))

plt.plot(X,Y,'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```
