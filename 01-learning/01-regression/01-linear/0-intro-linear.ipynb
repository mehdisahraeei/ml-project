```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

for example: y =  4x  + 4

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 2*(x) + 3 #formula

y_noise = 2 * np.random.normal(size=x.size) #for normlizing points
ydata = y + y_noise

plt.plot(x, ydata,  'go')
plt.plot(x,y, 'r--') 

plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```
