### importing libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### reading dataframe

```python
df = pd.read_csv('fuel.csv')
df.head(4)
```

```python
df['CYLINDERS'].value_counts
```

- stripplot , swarmplot

```python
# sns.stripplot(x = 'MAKE',y = 'ENGINESIZE', data = df, size = 15) #important problem: changing points 
# plt.figure().set_figwidth(15)
plt.figure(figsize=(15,7))
sns.swarmplot(x = 'MAKE',y = 'ENGINESIZE', data = df, size = 15)
plt.show()
```

- using hue: it shows components of on option with distinct colors

```python
# plt.figure().set_figwidth(15)
plt.figure(figsize=(15,7))
sns.swarmplot(x = 'MAKE',y = 'ENGINESIZE',hue = 'VEHICLECLASS',data = df, size = 15)
plt.show()
```

- jointplot: it shows histograms too

```python
# plt.figure().set_figwidth(15)
plt.figure(figsize=(15,7))
sns.jointplot(x = 'MAKE',y = 'ENGINESIZE', data = df, kind='scatter')
plt.show()
```

### pairplot: more detailed
- plot_kws={'s':80}: size
- palette: sets of colors

```python
sns.pairplot(data = df,hue='MAKE',palette='hls', plot_kws={'s':80})
plt.show()
```

### histplot

```python
sns.histplot(df['ENGINESIZE'])
```

# displot
- Seaborn distplot lets you show a histogram with a line on it. This can be shown in all kinds of variations.

```python
sns.displot(data=df,x='MAKE',y='CYLINDERS')
```

### heatmap
- In general, this code allows you to check the degree of correlation between fields
- if be 1 both variables are relatively similer to each other otherwise nope

```python
plt.figure(figsize=(14,9))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Greens')
plt.show()
```
